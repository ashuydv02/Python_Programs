# Merging two Series or Dataframes
join():- Join DataFrames using indexes.
concate():- Concatenate pandas objects along a particular axis.
merge():- Merge DataFrames by indexes or columns. 

```python
tempdf = {
    'Sno.' : ['10', '11'],
    'Name' : ['Ayush', 'Himanshu'],
    'Age' : [23, 24],
    'Salary' : [5000, 20000],
    'DoB' : ['2001-06-14', '2000-04-23'],
}
temp_df = pd.DataFrame(tempdf)
pd.concat([df, temp_df])
```

```python
left = pd.DataFrame({"key": ["foo", "foo"], "lval": [1, 2]})
right = pd.DataFrame({"key": ["foo", "foo"], "rval": [4, 5]})
pd.merge(left, right, on="key")
```

```python
left = pd.DataFrame({"key": ["foo", "bar"], "lval": [1, 2], 'col': ['1', '2']})
right = pd.DataFrame({"key": ["foo", "bar"], "rval": [4, 5], 'col1': ['4', '5']})
pd.merge(left, right, on="key")
```

```python
left.join(right, lsuffix='_left', rsuffix='_right')
left.set_index('key').join(right.set_index('key'))
```

# Grouping

```python
df1 = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)
print(df1)
df1.groupby(['A', 'B'])[['C', 'D']].sum()
```
